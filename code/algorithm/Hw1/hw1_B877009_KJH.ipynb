```python
def Fibonacci(input) :
  result = 1;

  n_1 = 1;
  n_2 = 1;

  for i in range(1, input+1):
    if i == 1 or i == 2 :
      result = 1;
    else :
      result = n_1 + n_2;

      n_2 = n_1
      n_1 = result
      

    print("i가" , i , "일 때 피보나치 수열의 값은 ", result, "이다.")


Fibonacci(40)
```

하노이의 탑 규칙 n번쨰의 블록을 움직이기 위해서

n이 1일경우는 1번만 움직이므로 1이다

n이 1이 아닐경우는 

def 하노이(n)
  1. n-1번째까지의 블록을 움직인다
  2. n번째 블록을 움직인다
  3. n-1번째까지의 블록을 다시 n번째 블록으로 움직인다

의 구조를 띄고 있다

1.인 n-1번쨰 블록을 움직인다를 실행하려면 n-1보다 위에있는 블록도 움직여야 하므로 하노이(n-1)와 같기 때문에

def 하노이(n)
  1. 하노이(n-1)
  2. n번째 블록을 움직인다
  3. 하노이(n-1)

가 되고 이를 다시 1. 하노이(n-1)를 실행하면

def 하노이(n-1)
  1. n-2번째까지의 블록을 움직인다
  2. n-1번째 블록을 움직인다
  3. n-2번째까지의 블록을 다시 n-1번째 블록으로 움직인다

가 실행되고 또다시 1.을 실행하면

def 하노이(n-2)
  1. n-3번째까지의 블록을 움직인다
  2. n-2번째 블록을 움직인다
  3. n-3번째까지의 블록을 다시 n-2번째 블록으로 움직인다

가 된다 결국 이를 반복하면

def 하노이(2)
  1. 1번째까지의 블록을 움직인다
  2. 2번째 블록을 움직인다
  3. 1번째까지의 블록을 다시 2번째 블록으로 움직인다

가 될때까지 반복 되는데 이 경우는 블록 2개를 움직일 때 이동횟수 3이다

3.같은 경우는 옮기는 블럭의 수는 같고 옮기는 위치만 다르므로 1.과 같은 횟수이다


2.는 n블록 1개만 움직이므로 1회이다

움직이는 개 3개 일경우

  1. 하노이(2)
  2. 2번째 블록을 움직인다
  3. 하노이(2)

가 되는데 3+1+3 = 7이다

움직이는게 4개 일 경우 7+1+7 = 15
움직이는게 5개 일 경우 15+1+15 = 31

결국 하노이(n) = 2^n-1 이다

10개를 옮기려면 511+1+511 = 1023이다



```python
def Hanoi(number, a, b, c ,result = 0):

    if number == 1:
        print(a, "번 위치의 블록을", c, "번 위치로 이동")
        return 1

    else:
        result2 = result

        result += Hanoi(number-1, a, c, b, result2)

        print(a, "번 위치의 블록을", c, "번 위치로 이동")
        result += 1

        result += Hanoi(number-1, b, a, c, result2)

        return result



print(Hanoi(10, 1, 2, 3), "번 이동했습니다")

```
